# for 반복문

- `for` 반복문은 iterable 객체의 요소를 하나씩 꺼내 실행한다.
- `str`, `list`, `tuple`, `dict`, `set`, `range` 등이 대표적인 iterable 객체이다.
- 반복 횟수가 정해져 있거나 컬렉션의 모든 요소를 순회할 때 자주 사용한다.

## 1. 문자열, 리스트, 튜플 순회

문자열은 문자 단위로, 리스트와 튜플은 요소 단위로 순회한다. (시퀀스 타입)

```python
text: str = 'hello world'
for ch in text: # text를 순회하면서 하나씩 꺼내어 ch에 삽입
    print(ch)

lst: list[int] = [10, 20, 30]
for num in lst:
    print(num)

tpl: tuple[int] = (30, 35, 40)
for num in tpl:
    print(num)

```

## 2. dict와 set 순회

`dict`를 그대로 순회하면 key가 나온다. `items()`를 사용하면 key와 value를 함께 꺼낼 수 있다.

`set`은 순서가 없는 자료형이므로 출력 순서는 실행 환경에 따라 달라질 수 있다.

## 3. range 객체

`range(start, end, step)`은 정수 범위를 표현하는 객체이다.

- `start`: 시작값
- `end`: 끝값, 포함되지 않음
- `step`: 증가 또는 감소 간격

## 4. 중첩 반복문

반복문 안에 반복문을 넣으면 행과 열처럼 2차원 구조를 만들 수 있다. 구구단과 별찍기 같은 예제에 자주 사용한다.

```python
print("---중첩 반복문 예시---")
for a in range (1,4):
    for b in range(1,3):
        print(a, b)
```

```python
print("---2단 출력하기 ---")
dan: int =2
for num in range (1,10):
    print(f'{dan} x {num} = {dan * num}')
```

```python
print("---구구단 전체 출력하기(2단 ~ 9단) ---")
for dan in range (2,10):
    for num in range (1,10):
        print(f'{dan} x {num} = {dan * num}\t', end='')
    print("-" * 30)
```

```python
print("---range를 이용한 반복 횟수 설정---")
print("🐙3행 5열 출력")

for row in range(3):
    for col in range(5):
        print('🐙', end='') # end 속성 : print() 종류 후 출력할 문자 지정
    print() #줄바꿈
```

## 5. zip과 enumerate

- `zip()`은 여러 iterable을 같은 순서끼리 묶는다.
- `enumerate()`는 인덱스와 요소를 함께 꺼낸다.

```python
print("---zip객체---")
names: list[str] = ['a', 'b','c']
nums: list[int] = [10, 20, 30]

z = zip(names, nums)
print(z, type(z)) # 반복문과 함께 써야함

for name, num in z:
    print(name, num)

print("---enumerate객체---")
foods: list[str] = ['제육', '피자', '훠궈', '치킨']
#숨겨진 인덱스 번호, 묶어서 반환.0-제육,1-피자,2-훠궈,3-치킨
for index, food in enumerate(foods):
    print(f'{index}번 인덱스: {food}')
```

## 6. 반복문으로 평균 구하기

반복문을 사용하면 리스트나 딕셔너리 리스트에 들어 있는 값을 누적해서 평균을 계산할 수 있다.

```python
print("---list에 담긴 값 평균 구하기---")
scores: list[int] = [20, 43 ,91, 49, 60]

score_sum = 0
for score in scores:
    score_sum += score # 합계 누적

score_avg = score_sum / len(scores) # 합계/값 개수 = 평균
print("평군: ", score_avg)
print("평균: ", sum(scores) / len(scores))  #숫자로 이루어진 단순 list인 경우에만 가능, (list, set, tuple)
```

```python
print("---list[dict]에서 점수 평균 구하기---")

students: list[dict[str, object]] = [
    {'name': '홍길동', 'score': 100},
    {'name': '신사임당', 'score': 77},
    {'name': '이순신', 'score': 85},
]

# students[0]['score']
# students[0].get('score')

sum: int = 0

# list 순회 이용
#for std in students:
#    sum += std.get('score') #std는 0,1,2 로 변경됨

# range+ indexing이용
for index in range(len(students)): # 0,1,2가 index에 순서대로 저장될 것임
    sum += students[index].get('score')


avg = sum / len(students)
print("평균: ", round(avg, 3)) # 반올림
```

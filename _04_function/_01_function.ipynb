# Function 함수

- 함수는 반복되거나 절차적인 실행 코드를 하나의 이름으로 묶은 것이다.
- 함수를 선언한 뒤 호출해서 사용한다.
- 호출할 때 값을 전달할 수 있고, 실행 결과를 반환받을 수 있다.
- Python에서 함수는 하나의 값처럼 다룰 수 있는 1급 객체이다.

## 1. 함수 선언과 호출

`def` 키워드로 함수를 선언하고, 함수 이름 뒤에 괄호를 붙여 호출한다.

```python
from IPython.utils.PyColorize import pride_theme


def add(a, b): # 이 함수가 실행될 때에는 값을 2개 전달해야한다.
    return a + b
print("add()함수 호출: ", add(1,2))  # add 함수 호출 ( call )
print("type(add): ", type(add))  #function
print(add) # 함수 객체(선언, 정의) 참조 == 함수 모양 그대로

# 값 전달 O, 결과 반환 X
def minus(a, b):
    print("minus()함수 호출:", a-b)

minus(3,1) # 함수 호출하여 값 출력

# 값 전달 X, 결과 반환 O
def multiply():
    return 111*111

print(multiply()) # 출력 문으로 호출하여 반환값 출력

# 값 전달 X, 결과 반환 X
def greeting():
    print("hello world")

greeting() # 호출만 하면 됨
```

## 2. 매개변수와 인수

- **매개변수(parameter)**: 함수 선언부에서 값을 받기 위해 만든 변수
- **인수(argument)**: 함수 호출부에서 실제로 전달하는 값

```python
def foo(k, m): # k, m == 매개변수(파라미터)
    print("k:" ,k)
    print("m:" ,m)

foo(100, 200)
```

## 3. return과 함수 객체

함수는 실행을 마치면 호출한 위치로 돌아온다. 이때 `return`으로 값을 하나 돌려줄 수 있다.

또한 Python에서 함수는 변수에 담거나 다른 함수의 인수로 전달할 수 있다. ( 1급 객체 )

### 1급 객체
- 값처럼 다룰 수 있는 객체
1. 변수에 대입 가능
2. 함수 인자로 전달 가능
3. 함수 반환 값으로 사용 가능
4. 자료 구조에 저장 가능

```python
# 일반적인 데이터 타입 return
def bar():
    print("bar")
    return 0
# bar 함수를 func 변수에 대입
func = bar # 가능 -> 1급 객체
print("bar 주소:", id(bar))
print("func 주소:", id(func))
print("bar is func:", bar is func) # 같은지 비교
# 주소가 같다 -> 주소만 복사, 얕은 복사와 유사

# 함수를 인자로 전달
def runner(f):
    print( f() )# 파라미터(매개변수)에 담긴 함수를 호출

runner(func) #러너 함수 호출-> func 함수 호출 -> bar 출력 -> 0 반환

# return 꼭 필요한지?
```

## 4. 위치 인수와 키워드 인수

함수를 호출할 때 인수를 전달하는 방식은 크게 두 가지이다.

- 위치 인수: 작성한 순서대로 매개변수에 전달
- 키워드 인수: `key=value` 형태로 같은 이름의 매개변수에 전달

두 방식을 섞을 때는 위치 인수를 먼저 쓰고, 그 뒤에 키워드 인수를 써야 한다.

```python
def test(a, b):
    print("a:", a)
    print("b:", b)
print("---위치 인수---")
test(10, 20) # 차례대로 10은 a 에  20은  b 에 들어갈 것이다 -> 위치 인수

print("---키워드 인수---")
test(b = 30, a =40) # 명시해준 파라미터에 대입될 것이다. -> 키워드 인수

print("---위치 + 키워드 인수---")
# test(b = 100, 200) => 키워드 인수 뒤에는 위치 인수를 작성할 수 없다. 무조건 위치 인수가 먼저 와야 함.
test(100, b = 200)


```

## 6. packing: `*args`, `**kwargs`

- `*args`: 여러 위치 인수를 하나의 튜플로 모은다.
- `**kwargs`: 여러 키워드 인수를 하나의 딕셔너리로 모은다.
- 매개변수 작성 순서는 일반 매개변수, `*args`, `**kwargs` 순서를 기본으로 생각하면 좋다.

```python
def test3(*args): # 전달 받은 인자를 tuple로 묶음
    print(args)
    return sum(args) # list,set,tuple에 저장된 모든 값을 더해서 반환

print(test3(1, 2, 3))  # 전달 받은 1, 2, 3 인자를 하나의 튜플로 반환
print(test3(1, 2, 3, 4, 5))

# 기본 인수 + *args
def test4(emoji, *args):
    print(emoji, args)

test4('🎉',1,2,3,'🎉') # 첫번째 emoji매개변수로 받은 것 제외하고 튜플로 따로 출력됨
test4('🎉'*3, 1,2,3,4,5)
test4('🎉') # 전달하지 않은 인자가 있을 시 비어있는 튜플 반환
```

```python
def test5(**kwargs): # 딕셔너리는 키와 밸류값으로 2개의 값이 있기 때문에 * 이 두개
    print(kwargs, type(kwargs))

test5() # dict 형식 확인
test5(menu="돈까스", price=10000, options=['치즈', '콜라']) # key:value 형식으로 만들어짐

# 기본 인수 + **kwargs (??실행 확인)
def test6(name, age, **kwargs):
    print(name, age, kwargs)

test6('홍길동', 20,a = 10, b = 30, hobby = ['game', 'study'])
#test6('홍길동', 20, a = 10, b = 30, ['game', 'study']) # **kwargs는 무조건 일반 인수 뒤에만 작성 가능

# 일반인수 + *args + **kwargs
def test7(a,b, *args,**kwargs):
    print('a,b: ', a,b)
    print('args: ', args)
    print('kwargs: ', kwargs)

# 호출 시 인수 순서: 일반 -> *args -> **kwargs
test7(10, 20, 30, 40, 50, name='이순신', age=50)
```

## 7. unpacking: `*list`, `**dict`

- `*list`는 리스트의 요소를 꺼내어 위치 인수처럼 나열한다.
- `**dict`는 딕셔너리의 key-value를 꺼내어 키워드 인수처럼 나열한다.

```python
names = ['홍길동', '이순신', '카리나']

def test8(*args):
    print(args)

print("---list unpacking---")
test8(names) # tuple[list]형태로 저장됨 -> list로 묶인 상태
test8(names[0],names[1],names[2]) # tuple[str,str,str]
test8(*names) # * 의 의미: 묶거나 풀거나.
```

```python
info ={
    'date': '2026/06/04',
    'time': '14:06'
}

def test9(date, time):
    print('date:', date)
    print('time:', time)

test9(info['date'], info['time']) # 위치인수
test9(time=info['time'], date=info['date']) # 키워드 인수
```

## 8. 위치 전용과 키워드 전용 매개변수

Python 함수는 특정 매개변수를 위치 전용 또는 키워드 전용으로 제한할 수 있다.

- `/` 앞의 매개변수는 위치 전용이다.
- `*` 뒤의 매개변수는 키워드 전용이다.

```python
def test10(name,/, greeting="안녕"): # / 기준 왼쪽은 위치 전용 인수
    print(name, greeting)

test10("홍길동")
# test10(name="홍길동", greeting="hello") # / 기준 왼쪽은 위치 전용 인수이기 때문에 타입 에러 뜸
# -> / 앞 name은 위치 인수로만 동작


#def test11(a, b, c):
#    print(a ,b, c)
#test11(10, 20, 30) # 전부 위치 인수로 적용
def test11(a, * , b, c):
    print(a ,b, c)

# test11(10, 20, 30) # Error
# - * 뒤는 키워드 인수로만 동작
test11(10, b = 20, c = 30)
```

## 9. 함수 type hint

함수의 매개변수와 반환값에도 type hint를 작성할 수 있다. type hint는 실행을 강제하기보다 IDE와 정적 분석 도구가 코드를 이해하도록 돕는 역할을 한다.

```python
# 매개 변수 타입 힌트 => a: str, b: int, c: bool
# 반환형 타입 힌트 => " -> dict[str, object] "
def test12(a: str, b: int, c: bool) -> dict[str, object]:
    return {'a':a, 'b':b, 'c':c}

print(test12('홍길동', 20, True))
```

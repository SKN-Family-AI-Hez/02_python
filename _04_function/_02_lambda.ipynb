# lambda 람다

- `lambda`는 함수를 한 줄로 간결하게 선언하는 문법이다.
- 이름이 없는 작은 함수를 만들 때 사용한다.
- 기본 구조는 `lambda 매개변수: 반환값`이다.
- 복잡한 로직은 일반 `def` 함수로 작성하는 편이 읽기 쉽다.
```
변수명 = lambda (매개변수): (함수 실행 내용)
# 함수 실행 내용에 작성된 코드(표현식)이 끝나면 자동으로 반환된다
```

## 1. lambda 기본 문법

아래 두 코드는 같은 의미이다.

```python
def square(x):
    return x ** 2

square = lambda x: x ** 2
```

````python
def square(x: int) -> int:
    return x ** 2

squeare2 = lambda x : x ** 2 # square  함수를 square2에 대입

print(square(2))
print(squeare2(2))
````

## 2. list comprehension 안에서 사용하기

`lambda`는 짧은 계산식을 값처럼 전달하거나 즉시 실행할 때 사용할 수 있다.

````python
def make_squares() -> list[int]:
    return [(lambda x: x ** 2)(n) for n in range(1,11)]# 1~10까지의 숫자를 하나씩 꺼내어 n에 저장하고 lambda식 연산 후 리스트로 출력
print('1~10 제곱: ', make_squares())
````

## 3. 여러 매개변수 사용

`lambda`도 일반 함수처럼 하나 이상의 매개변수를 받을 수 있다.

````python
add = lambda a, b: a + b
# 함수이름 def 매개변수 return a*b

print(add(10, 20))
````

## 4. 기본값, `*args` 사용

`lambda`에서도 기본값이 있는 매개변수와 `*args`를 사용할 수 있다. 다만 코드가 길어지면 `def` 함수로 분리하는 것이 좋다.

````python
print("---lambda 매개변tn 기본값---")

#multiply = lambda a, b: a * b
#print(multiply(10)) # a에 10이 대입되고 b 가 없으니 에러 남

multiply = lambda a, b=1: a * b # 매개변수 기본 값 1로 설정
print(multiply(10))

print("---lambda *args---")
adder = lambda *args: sum(args) #전달되는 모든 형태를 튜플형태로 묶어서 반환    (?)
print(adder(1, 2, 3, 4, 5))
````

## 5. lambda와 고차 함수

고차 함수는 함수를 인자로 받거나, 함수를 반환하는 함수이다.

`lambda`는 이름 없는 함수를 값처럼 만들어서 다른 함수에 전달할 때 자주 사용한다.
아래 예제에서 `apply_to_numbers()`는 `func`라는 함수를 인자로 받는 고차 함수이다.


````python
numbers = [1, 2, 3, 4, 5]

# 매개변수 func == 나중에 실행된다, 콜백 함수
# apply_to_numbers == 고차함수

def apply_to_numbers(numbers, func): # 매개변수를 주면 result 리스트를 반환해주는 함수
    result = []

    for n in numbers: # 전달받은 넘버즈 만큼 연산
        result.append(func(n)) # 전달받은 함수를 실행해서 반환되는 값을 리스트에 추가한다.
    return result
# ---------여기까지 정의

numbers1 = apply_to_numbers(numbers, lambda n : n * 2) # 호출
print(numbers1)

# numbers와 lambda 를 매개변수로 해서 apply 함수 호출, for문 실행, func(lambda)실행 후 리스트 생성, 출력

numbers2 = apply_to_numbers(numbers, lambda n : n * 3)
print(numbers2)

numbers3 = apply_to_numbers([n for n in range(2, 11, 2)], lambda n: n-1) # 2,4,6,8,10을 전달해서 함수(n-1) 실행
print(numbers3)
````

````python
# 함수를 반환하는 고차 함수
def make_multiplier(n: int):
    return lambda number: number * n

triple = make_multiplier(3) # lambda number: number * 3
double = make_multiplier(2)
print(triple(10)) # 10 이 number로 입력되어 return 10 * 3
print(double(10))
````

## 6. lambda와 콜백 함수

콜백 함수는 내가 바로 실행하는 함수가 아니라, 다른 함수에 전달해 두었다가 필요한 시점에 호출되는 함수이다.

`lambda`는 콜백 로직이 짧을 때 사용하기 좋다. 다만 콜백 안의 코드가 길어지면 이름 있는 `def` 함수로 작성하는 편이 읽기 쉽다.


````python
# 매개변수 operation == 콜백함수
# calculator == 고차함수
def calculator(a, b, operation):
    return operation(a, b)

add_result = calculator(999, 888, lambda x, y: x + y)
print("add_result: ", add_result)

minus_result = calculator(999, 888, lambda x, y: x - y)
print("minus_result: ", minus_result)
````

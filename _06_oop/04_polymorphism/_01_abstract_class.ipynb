# 추상클래스 Abstract Class
- 자식클래스에서 구현해야 할 추상메소드를 가진 부모클래스 (인터페이스)
    - 추상메서드 == 미완성 메서드 == 메서드 정의 부분이 비어 있음

- 추상메소드는 자식클래스에서 반드시 구현(오버라이딩)해야할 의무를 가진다.
- 추상클래스는 직접 객체화 할수 없다.( 추상 클래스로 인스턴스 생성 불가 )


- abc.ABC클래스를 상속하면 추상클래스를 선언할 수 있다.
    - ABC == Abstract Base Class
- @abstractmethod 데코레이터로 추상메소드를 선언할 수 있다.


- class: 인스턴스를 만들기 위한 속성/기능을 정의한 것 (설계도)

```python
from abc import ABC, abstractmethod

# 추상 클래스 정의
# - ABC 상속 == 해당 클래스가 너무 추상적 또는 해당 클래스로 인스턴스 생성을 하지 않겠다는 의미

# 추상 클래스의 목적
# - 상속받은 자식 클래스들에게 공통된 규약(규칙)을 제공

class Product(ABC):

    def __init__(self, code: str, name: str, price: int):
        self.__code = code        # private
        self.__name = name        # private
        self.__price = price      # private


    #!!!!!!1@ 의미
    # 추상 메서드 == 미완성 메서드 == 정의 부분이 없음
    @abstractmethod      # 추상 메서드임을 표시
    def get_tax(self):
        pass           # 정의 부분은 상속받은 자식에게 넘김

    def get_price(self):
        return self.__price  #!!!!return priint차이

    def __str__(self): #!!!! 기본구조
        return f'code={self.__code}, name={self.__name}, price={self.get_price()}'
```

```python
# 추상클래스로 인스턴스 생성하기 -> 오류 확인
# prod = Product("111", "노트북", 2000000) # 추상클래스는 인스턴스화 할 수 없음.

```

```python
# 추상 클래스를 상속 받은 자식 클래스
class TV(Product):  # alt + enter 로 기본 구조 잡기, 메서드 구현

    def __init__(self, code: str, name: str, price: int, size: int):
        super().__init__(code, name, price)
        self.__size = size

    # 상속 받은 추상 메서드를 재작성 == 오버라이딩
    def get_tax(self): # 부가세를 10%로 책정하여 반환
        return self.get_price() * 0.1

    # __str__() 오버라이딩
    def __str__(self):
        return f'code={super().__str__()}, size={self.__size}'

```

```python
tv1 = TV("samsung-tv-1234", "SAMSUNG OLED TV", 5_000_000, 85)
print(tv1)
print(tv1.get_tax())

tv2 = TV("lg-tv-1234", "LG Flexible TV", 10_000_000, 100)
print(tv2)
print(tv2.get_tax())
```

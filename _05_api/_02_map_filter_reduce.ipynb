## map
iterable객체의 각요소에 대해 지정한 함수를 적용하고, 그 결과를 map객체로 반환
```
map(function, iterable) -> map
```




````python
from debugpy.common.timestamp import current

nums: list = [1,2,3,4,5]

result = map(lambda x : x ** 2, nums)
print(result, type(result)) # map 주소 , map 타입
print("list(result):", list(result))

# map(): 기존 list의 요소를 하나씩 꺼내 함수를 호출, 함수 결과들을 모아 새로운 list를 생성
````

````python
# 문자열 리스트 -> 문자열 길이 리스트
````

````python
products: list[str] = ['아이스 아메리카노', '수박주스', '말차라떼', '식혜', '딸기 스무디']

products_len: list[int] = map(lambda x: len(x), products) # x 에 products의 요소들이 삽입됨
print(list(products_len))  # 이렇게 하지 않으면 map 타입으로 보임
````

````python
# 여러 순회 가능 객체 (list) 통합
mid_term = [90, 55, 70]
final_term = [100, 40, 75]

avg_scores = list( map( lambda x, y : (x + y) / 2 , mid_term, final_term )) # map(콜백함수, midterm, finalterm)
print(avg_scores)
````

## filter

주어진 iterable객체에 대해서 참/거짓 평가후 참인 요소들만 결과값으로 반환

```
filter(function, iterable) -> filter
```

````python
nums = list(range(1,11))

evens = filter(lambda n: n % 2 == 0, nums) # nums 리스트 내에서 짝수만 뽑는 필터
odds = filter(lambda n: n % 2 != 0, nums) # 홀수
threes = filter(lambda n: n % 3 == 0, nums) # 3의 배수

print(evens, type(evens))
print("list(evens):", list(evens))
print("list(odds):", list(odds))
print("list(threes):", list(threes))
````

````python
# 이름 목록에서 성이 "홍" 씨인 사람만 찾아서 list로 반환하기
names: list[str] = ['홍길동', '고길동', '홍징경','최홍만','홍명보']

# str.startswith(str2) -> bool
# - 문자열의 시작이 str2와 같다면 True
hong_faily = list(filter(lambda name: name.startswith('홍'), names)) # list(filter(콜백함수, names))
print(hong_faily)
````

````python
# dict 검색(dict 필터링)
# 학생 목록에서 점수가 90 점 이상인 학생만 필터링
students = [
    { 'name' : '이순신', 'score' : 80},
    { 'name' : '윤봉길', 'score' : 75},
    { 'name' : '논개', 'score' : 99},
    { 'name' : '강감찬', 'score' : 91},
]

# filter()함수
students1 = list(filter(lambda s: s['score'] >= 90, students))
print(students1, type(students1))

# list comprehension
student2 = [s for s in students if s['score'] >= 90]  # for문 실행 > if 조건문 실행 >  s에 저장
print(student2)
````

## reduce
iterable객체의 요소에 대한 연산결과를 단일값을 반환.
합계, 최대/최소값 등의 연산에 적합

```
reduce(function, iterable, initial_value) -> Any
```
함수는 3개의 인자를 받는다.
1. 누적 연산값
2. 현재 요소
3. 초기값 (선택)



````python
#!!!!!!

from functools import reduce      # 외부 functools 모듈에서 reduce 함수만 가져오기

# 정수형 list 내 요소들의 합계
nums: list[int] = [ 11, 22, 33, 44, 55, 66, 77, 88, 99]

# 총합 저장
#total =  reduce(콜백함수(누적용변수,현재값), 순회 객체(iterable), 누적변수 초기 값)
# -> 콜백 함수의 return 값이 자동으로 누적용 변수에 대입 된다
total =  reduce(lambda sum, curr: sum + curr, nums, 0) # 초기값 0인 이유: 더하거나 빼도 영향이 없도록. 곱셈인 경우에는 1로 초기화 고려.
print(total)
````

````python
# 최대, 최소 값 찾기
nums: list[int] = [ 32,123,321,-43,3,-12345,1111]

# 최댓값 찾기
#def func(max, curr):
#    if curr < max:
#        return max
#    else :
#        return curr

max_num: int = reduce(lambda max, curr: max if max > curr else curr, nums, 0)
print(max_num)

# 최솟값 찾기
min_num: int = reduce(lambda min, curr: min if min < curr else curr, nums, 0)
print(min_num)

# 절대 값이 가장 큰 수 + 일반 함수
def abs_func(max, curr):
    print(f'max: {abs(max)}, curr: {abs(curr)}')
    return max if abs(max) > abs(curr) else curr

abs_max = reduce(abs_func, nums, 0) # 함수를 매개변수로 그대로 집어넣음.
print("abs_mas:",abs_max)
````
